```python

#Inner JoinLeft JoinRight Join Outer Join
import pandas as pd

# Creating DataFrame 1
data1 = {
    "StudentID": [101, 102, 103, 105],
    "Age": [20, 21, 22, 19],
    "Grade": ["A", "B", "A", "C"]    
}
df1 = pd.DataFrame(data1)  # Convert dictionary to DataFrame

# Creating another DataFrame (example)
data2 = {
    "StudentID": [101, 102, 104, 105],
    "Marks": [85, 78, 90, 65]
}
df2 = pd.DataFrame(data2)

# Merging DataFrames
df3 = pd.merge(df1, df2, on="StudentID", how="inner")  # Use 'left', 'right', or 'outer' as needed

print(df3)
```

```python
#left join
df3 = pd.merge(df1, df2, on="StudentID", how="left")  # Use 'left', 'right', or 'outer' as needed
df3
```

```python
#rightjoin
df3= pd.merge(df1,df2, on="StudentID", how="right")
df3
```

```python
#outer join
df3=pd.merge(df1,df2,  on="StudentID", how="outer")
df3
```
